```python
#import library's we are using

from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#get the http request and load it into beautifulsoup

req = requests.get("https://ehealth88.nl/over-ons/het-team/")
html = req.text
soup = BeautifulSoup(html, "lxml")
#with open("hetteam.html") as f:
    #soup = BeautifulSoup(f, "lxml")
```

```python
#select the list with the person names and links

menu = soup.select(".mk-grid ul .team-info-wrapper a")
```

```python
#loop through the list and get both names and links in a dict while also cleaning the data

menu_list = []
for menu_item in menu:
    item = menu_item.get_text().replace("\n", " ").strip()
    link = menu_item.get("href")
    person ={
        "aname" : item,
        "link" : link
    }
    menu_list.append(person)
```

```python
#export dict to dataframe and show the dataframe

df = pd.DataFrame(menu_list)
df
```

```python
#export the dataframe to csv

df.to_csv("employ.csv")
```

```python
#loop through the links and get the description, clean the description and add to the exsisting dict

index = 0
for url in menu_list:
    req_loop = requests.get(menu_list[index]["link"])
    html_loop = req_loop.text
    soup_loop = BeautifulSoup(html_loop, "lxml")
    description_select = soup_loop.select("#text-block-10")[0].get_text().replace("\n", " ").strip()
    menu_list[index].update({"des" : description_select})
    index = index + 1
```

```python
#create a updated dataframe and display it

df2 = pd.DataFrame(menu_list)
df2
```
