```python
#importing the right library's
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import numpy as np
```

```python
#loading in the dataset, we have to set header to none, because the datafile has no collumn names
df = pd.read_csv("Assignment 3 - data jokes.csv", header = None)
df_input = pd.read_csv("Assignment 3 - data jokes.csv", header = None)
df.head()
```

```python
#here we replace the 99s with nans and remove all the collumns which contains Nans, so we have valuable data we can put into our model.
df = df.replace(99.0, np.nan)
df = df.dropna(axis = 1)
df.head()
```

```python
#here we fit our K-mean model, we use all the collumns from our previous dataframe. The last line will make a new collumn with the cluster per person.
km = KMeans(n_clusters=6, random_state=1) 
X = df.iloc[:, 0 : 6] 
km = km.fit(X)
df["cluster"] = km.predict(X) 
```

```python
#make a pivot table of our predictions so we can analyse our clusters
df_subset = pd.pivot_table(data=df, index = "cluster", aggfunc = 'mean')
df_subset
```

```python
#Here we make two loops, one that anylises which joke numbers are in our current dataframe. The second one print every joke from the text thats in our Dataframe
jokes_text = open("Assignment 3 - text jokes.txt")
jokes_indexes = []
index = 1 #index starts with 1 because the text joke file starts with 1
for collumn in df_subset:
    jokes_indexes.append(collumn)
for line in jokes_text:
    if index in jokes_indexes:
        print(line)
    index += 1
```

```python
#define the function which does 2 things, it takes the number input and correlates it with the dataframe.
#after this it sorts with the highest correlation first.
def joke_corr(jokenumber):
    df_corr = df_input.corrwith(df_input[int(jokenumber)]).sort_values(ascending = False)
    return df_corr

#here we ask for a input number and call the function with it
joke_ask = input("What number between 1-149 you want to check the correlations with?")
df_joke = joke_corr(joke_ask)
df_joke.head()
```
